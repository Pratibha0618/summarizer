```python
vali_df= pd.read_csv('/kaggle/input/pubmed-article-summarization-dataset/validation.csv',encoding='latin-1' )
```

```python
vali_df
```

```python
vali_df= vali_df.rename(columns={'abstract':'target_text', 'article':'source_text'})
vali_df= vali_df[['source_text','target_text']]
```

```python
vali_df.head()
```

```python
vali_df['source_text']= 'summarize: ' + vali_df['source_text']
```

```python
vali_df.head()
```

```python
from simplet5 import SimpleT5

model= SimpleT5()
model.from_pretrained(model_type='t5', model_name='t5-base')
```

```python
# Ensure 'text' column contains strings
vali_df['source_text'] = vali_df['source_text'].astype(str)
vali_df['target_text'] = vali_df['target_text'].astype(str)
```

```python
vali_df.head()
```

```python
train_ds, test_ds= train_test_split(vali_df, test_size=0.3)
train_ds.shape, test_ds.shape
```

```python
model.train(train_df=train_ds[:5000],
            eval_df=test_ds[:100], 
            source_max_token_len=128, 
            target_max_token_len=50, 
            batch_size=8, max_epochs=5, use_gpu=True)
```
